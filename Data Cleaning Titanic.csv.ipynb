```python
import pandas as pd
import numpy as np
```

```python
#Mengimport data titanic
data = pd.read_csv('titanic.csv')
```

```python
#Melihat 100 data dari paling bawah
data.tail(100)
```

```python
#Menampilkan informasi column dari data titanic
data.info()
```

### Berdasarkan informasi column diatas, dapat dilihat bahwa terdapat beberapa data yang NaN dan adanya column yang terisi full NaN.

### Oleh karena itu akan dilakukan proses delete column "Unnamed: 15" dikarenakan data yang full NaN, dan column "deck" dikarenakan terlalu banyak data NaN dan jikalau dilakukan proses drop dengan axis index akan menghilangkan data yang bagus.

```python
#Menghapus column Unnamed:15
data.drop('Unnamed: 15', inplace=True, axis=1)
```

```python
#Menghapus column deck
data.drop('deck', inplace=True, axis=1)
```

```python
#Melihat hasil data yang telah dihapus columnnya
data
```

```python
data.info()
```

```python
#Menampilkan unique value dari setiap column yang ada
for column in data.columns:
    print(" ")
    print("Unique Values from this", column," is", data[column].unique())
```

```python
#Melihat banyaknya data NaN dari setiap column
data.isnull().sum()
```

```python
#Menghapus semua data yang memiliki data NaN
data.dropna(inplace = True)
```

```python
#Melihat hasil drop data NaN
data.info()
```

```python
#Memeriksa kembali data titanic, apakah masih ada data NaN atau tidak
data.isnull().sum()
```

```python
data.to_csv("Titanic_Clean.csv", index = False)
```
